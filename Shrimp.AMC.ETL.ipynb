```python
amc_csv_file = "AMC.csv"
amc_data_df = pd.read_csv(amc_csv_file)

shrimp_csv_file="shrimp-prices.csv"
shrimp_data_df=pd.read_csv(shrimp_csv_file)
```

##### Transform

```python
#Shrimp Data Clean up 

shrimp_data_df.rename(columns={'Month':'Date'}, inplace=True)
shrimp_data_df
shrimp_data_df["Date"]=pd.to_datetime(shrimp_data_df["Date"], errors='raise', dayfirst=False, yearfirst=False, utc=None, format=None, exact=True, unit=None, infer_datetime_format=False, origin='unix', cache=True)
shrimp_data_df

shrimp_data_df.rename(columns={'Change':'Percent Change'}, inplace=True)

shrimp_data_df["Percent Change"]=shrimp_data_df["Percent Change"]/100
shrimp_data_df
```

```python
#AMC data cleanup
amc_data_df
amc_data_df["Open"].pct_change
amc_data_df
```

```python
amc_data_df["Open_%_Change"]=amc_data_df["Open"].pct_change()
amc_data_df["High_%_Change"]=amc_data_df["High"].pct_change()
amc_data_df["Low_%_Change"]=amc_data_df["Low"].pct_change()
amc_data_df["Close_%_Change"]=amc_data_df["Close"].pct_change()
amc_data_df["Date"]=pd.to_datetime(amc_data_df["Date"])
amc_data_df
```

```python
#Merging the AMC and Shrimp Data
merged_amc_shrimp = pd.merge(shrimp_data_df, amc_data_df, on="Date")
merged_amc_shrimp.rename(columns={'Price':'Shrimp Price (per pound)'}, inplace=True)
merged_amc_shrimp.rename(columns={'Percent Change':'Shrimp Price Percent Change'}, inplace=True)
col=merged_amc_shrimp.columns
merged_amc_shrimp=merged_amc_shrimp[['Date', 'Shrimp Price (per pound)', 'Shrimp Price Percent Change',
       'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
#Removing AMC stock % change columns for simplicity 
merged_amc_shrimp
```

```python
#Cleaning up AMC 
amc_positive_df=amc_data_df.loc[amc_data_df["Open_%_Change"]>0,["Date", "Open", "Open_%_Change"]]
amc_negative_df=amc_data_df.loc[amc_data_df["Open_%_Change"]<0,["Date", "Open", "Open_%_Change"]]

amc_positive_df
amc_negative_df
```
